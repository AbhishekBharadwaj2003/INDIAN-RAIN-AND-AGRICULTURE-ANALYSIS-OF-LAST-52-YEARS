```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
df = pd.read_csv("rain-agriculture.csv")
```

```python
df["Total Rainfall"] = df[["JUN", "JUL", "AUG", "SEP"]].sum(axis=1)
```

```python
df.shape
```

```python
df.columns
```

```python
# Trend of Rice Production Over the Years (by State)

rice_trend = df.groupby(['Year', 'State Name'])['RICE PRODUCTION (1000 tons)'].sum().unstack()
rice_trend.plot(title='Rice Production Trend by State')
plt.ylabel('Rice Production (1000 tons)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#Average Rice Yield Per State

avg_rice_yield = df.groupby('State Name')['RICE YIELD (Kg per ha)'].mean().sort_values(ascending=False)
avg_rice_yield.plot(kind='bar', title='Average Rice Yield by State')
plt.ylabel('Rice Yield (Kg per ha)')
plt.show()
```

```python
wheat_production = df.groupby('State Name')['WHEAT PRODUCTION (1000 tons)'].sum().sort_values(ascending=False)

wheat_production.plot(kind='bar',color='orange', title='Total Wheat Production by State')

plt.ylabel('Wheat Production (1000 tons)')
plt.show()
```

```python
crop_yield_columns = ['RICE YIELD (Kg per ha)', 'WHEAT YIELD (Kg per ha)', 'MAIZE YIELD (Kg per ha)']
df_yield_trends = df.groupby('Year')[crop_yield_columns].mean()
df_yield_trends.plot(title='Crop Yield Trends Over Years')
plt.ylabel('Yield (Kg per ha)')
plt.show()
```

```python
corr_rain_yield = df[['Total Rainfall', 'RICE YIELD (Kg per ha)']].corr()
sns.heatmap(corr_rain_yield, annot=True, cmap='coolwarm')
plt.title('Correlation: Rainfall vs Rice Yield')
plt.show()
```

```python
rain_by_year = df.groupby('Year')['Total Rainfall'].mean()
top_rain_years = rain_by_year.sort_values(ascending=False).head(5)
low_rain_years = rain_by_year.sort_values().head(5)

top_rain_yield = df[df['Year'].isin(top_rain_years.index)].groupby('Year')['RICE PRODUCTION (1000 tons)'].mean()
low_rain_yield = df[df['Year'].isin(low_rain_years.index)].groupby('Year')['RICE PRODUCTION (1000 tons)'].mean()

top_rain_yield.plot(label='High Rainfall Years')
low_rain_yield.plot(label='Low Rainfall Years')
plt.legend()
plt.title('Rice Production in High vs Low Rainfall Years')
plt.ylabel('Rice Production (1000 tons)')
plt.show()
```

```python
major_crops = ['RICE PRODUCTION (1000 tons)', 'WHEAT PRODUCTION (1000 tons)', 'MAIZE PRODUCTION (1000 tons)']
crop_prod_by_state = df.groupby('State Name')[major_crops].mean().sort_values(by='RICE PRODUCTION (1000 tons)', ascending=False)
crop_prod_by_state.plot(kind='bar', title='Average Major Crop Production by State')
plt.ylabel('Production (1000 tons)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
oilseed_top5 = df.groupby('State Name')['OILSEEDS PRODUCTION (1000 tons)'].sum().sort_values(ascending=False).head(5)
oilseed_top5.plot(kind='bar', title='Top 5 States by Oilseeds Production')
plt.ylabel('Oilseeds Production (1000 tons)')
plt.show()
```

```python
rice_area_trend = df.groupby('Year')['RICE AREA (1000 ha)'].mean()
rice_area_trend.plot(title='Average Rice Cultivated Area Over Years')
plt.ylabel('Rice Area (1000 ha)')
plt.show()
```

```python
rain_cols = ['JUN', 'JUL', 'AUG', 'SEP']
monthly_rain = df.groupby('Year')[rain_cols].mean()
monthly_rain.plot(title='Monthly Rainfall Trend Over Years')
plt.ylabel('Rainfall (mm)')
plt.show()
```

# FORECASTING

# 1. Overall Wheat Production Forecast 

```python
# Aggregate total wheat production per year
wheat_yearly = df.groupby('Year')['WHEAT PRODUCTION (1000 tons)'].sum().reset_index()
```

```python
import matplotlib.pyplot as plt
```

```python

plt.plot(wheat_yearly['Year'], wheat_yearly['WHEAT PRODUCTION (1000 tons)'], marker='o')
plt.title("Wheat Production Over the Years")
plt.ylabel("Wheat Production (1000 tons)")
plt.xlabel("Year")
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
# Create features
X = wheat_yearly['Year'].values.reshape(-1, 1)
y = wheat_yearly['WHEAT PRODUCTION (1000 tons)'].values
```

```python
# Train model
model = LinearRegression()
model.fit(X, y)
```

```python
# Forecast next 10 years
future_years = np.arange(wheat_yearly['Year'].max() + 1, wheat_yearly['Year'].max() + 11).reshape(-1, 1)
future_preds = model.predict(future_years)
```

```python
# Plot
plt.plot(X, y, label='Actual', marker='o')
plt.plot(future_years, future_preds, label='Forecast', linestyle='--', color='orange', marker='x')
plt.xlabel('Year')
plt.ylabel('Wheat Production (1000 tons)')
plt.title('Wheat Production Forecast (Linear Regression)')
plt.legend()
plt.grid(True)
plt.show()
```

# 2. STATE WISE FORECASTINGG

```python
from sklearn.linear_model import LinearRegression
import numpy as np

statewise_forecasts = {}

states = df['State Name'].unique()
years = df['Year'].unique()
max_year = years.max()

for state in states:
    state_data = df[df['State Name'] == state]
    yearly_data = state_data.groupby('Year')['WHEAT PRODUCTION (1000 tons)'].sum().reset_index()

    if len(yearly_data) >= 5:
        X = yearly_data['Year'].values.reshape(-1, 1)
        y = yearly_data['WHEAT PRODUCTION (1000 tons)'].values

        model = LinearRegression()
        model.fit(X, y)

        future_years = np.arange(max_year + 1, max_year + 11).reshape(-1, 1)
        predictions = model.predict(future_years)

        statewise_forecasts[state] = {
            'Year': future_years.flatten(),
            'Forecasted Wheat Production': predictions
        }
```

```python
state = 'assam'  

actual_data = df[df['State Name'] == state].groupby('Year')['WHEAT PRODUCTION (1000 tons)'].sum().reset_index()
forecast_data = statewise_forecasts[state]

plt.plot(actual_data['Year'], actual_data['WHEAT PRODUCTION (1000 tons)'], label='Actual', marker='o')
plt.plot(forecast_data['Year'], forecast_data['Forecasted Wheat Production'], label='Forecast', linestyle='--', color='orange', marker='x')
plt.title(f"Wheat Production Forecast: {state.title()}")
plt.xlabel("Year")
plt.ylabel("Wheat Production (1000 tons)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
state = 'west bengal'  

actual_data = df[df['State Name'] == state].groupby('Year')['WHEAT PRODUCTION (1000 tons)'].sum().reset_index()
forecast_data = statewise_forecasts[state]

plt.plot(actual_data['Year'], actual_data['WHEAT PRODUCTION (1000 tons)'], label='Actual', marker='o')
plt.plot(forecast_data['Year'], forecast_data['Forecasted Wheat Production'], label='Forecast', linestyle='--', color='orange', marker='x')
plt.title(f"Wheat Production Forecast: {state.title()}")
plt.xlabel("Year")
plt.ylabel("Wheat Production (1000 tons)")
plt.legend()
plt.grid(True)
plt.show()
```
